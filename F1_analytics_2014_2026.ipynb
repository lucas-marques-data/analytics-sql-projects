```python
# =============================================================================
# F1 ANALYTICS — COMO SERÁ 2026?
# Periodo: 2014-2026 | Bortoleto (Brasil) | Audi | Previsao 2026
# Execute tudo de uma vez para validar
# =============================================================================

# -----------------------------------------------------------------------------
# BLOCO 0: INSTALACAO E IMPORTS
# -----------------------------------------------------------------------------

# kaleido 0.2.1 — versao estavel para renderizacao de graficos no Colab/GitHub
!pip install kaleido==0.2.1 -q
!pip install requests pandas numpy plotly matplotlib seaborn -q

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import time
import warnings
warnings.filterwarnings('ignore')

# renderiza graficos Plotly como imagem estatica (necessario para exibir no GitHub)
pio.renderers.default = "png"

plt.style.use('default')
sns.set_palette('husl')

print("Bibliotecas carregadas.")
print("F1 ANALYTICS PROJECT — 2014-2026")

# -----------------------------------------------------------------------------
# BLOCO 1: COLETA DE DADOS HISTORICOS — API ERGAST (2014-2024)
# -----------------------------------------------------------------------------

print("\n" + "="*80)
print("BLOCO 1: COLETANDO DADOS HISTORICOS VIA API ERGAST...")
print("="*80)

def get_race_results(year):
    all_results = []
    try:
        url = f"https://api.jolpi.ca/ergast/f1/{year}/results.json?limit=1000"
        response = requests.get(url, timeout=15)
        if response.status_code != 200:
            print(f"   Aviso — Ano {year}: Status {response.status_code}")
            return pd.DataFrame()

        data = response.json()
        races = data['MRData']['RaceTable']['Races']

        for race in races:
            circuit = race.get('Circuit', {})
            for result in race.get('Results', []):
                driver      = result.get('Driver', {})
                constructor = result.get('Constructor', {})
                fastest_lap = result.get('FastestLap', {})

                all_results.append({
                    'year':                    int(race.get('season', year)),
                    'round':                   int(race.get('round', 0)),
                    'race_name':               race.get('raceName', ''),
                    'circuit_name':            circuit.get('circuitName', ''),
                    'country':                 circuit.get('Location', {}).get('country', ''),
                    'date':                    race.get('date', ''),
                    'driver_id':               driver.get('driverId', ''),
                    'driver_name':             driver.get('givenName','') + ' ' + driver.get('familyName',''),
                    'driver_nationality':      driver.get('nationality', ''),
                    'constructor_id':          constructor.get('constructorId', ''),
                    'constructor_name':        constructor.get('name', ''),
                    'constructor_nationality': constructor.get('nationality', ''),
                    'grid':                    int(result.get('grid', 0)),
                    'position':                pd.to_numeric(result.get('position', None), errors='coerce'),
                    'position_text':           result.get('positionText', ''),
                    'points':                  float(result.get('points', 0)),
                    'laps':                    int(result.get('laps', 0)),
                    'status':                  result.get('status', ''),
                    'fastest_lap_time':        fastest_lap.get('Time', {}).get('time', None),
                    'fastest_lap_rank':        pd.to_numeric(fastest_lap.get('rank', None), errors='coerce')
                })

        print(f"   {year}: {len(races)} corridas | {len(all_results)} registros")
        return pd.DataFrame(all_results)

    except Exception as e:
        print(f"   Erro em {year}: {e}")
        return pd.DataFrame()

print("Coletando 2014-2024. Aguarde ~3 minutos...\n")

all_seasons = []
for year in range(2014, 2025):
    df_year = get_race_results(year)
    if not df_year.empty:
        all_seasons.append(df_year)
    time.sleep(0.5)

df_historical = pd.concat(all_seasons, ignore_index=True)

print(f"\nDados historicos carregados.")
print(f"   Total: {len(df_historical):,} registros")
print(f"   Periodo: {df_historical['year'].min()}-{df_historical['year'].max()}")
print(f"   Pilotos: {df_historical['driver_name'].nunique()}")
print(f"   Equipes: {df_historical['constructor_name'].nunique()}")

# -----------------------------------------------------------------------------
# BLOCO 2: DADOS 2025 + TREINOS 2026
# -----------------------------------------------------------------------------

print("\n" + "="*80)
print("BLOCO 2: ADICIONANDO DADOS 2025 E TREINOS 2026...")
print("="*80)

# Tentar API para 2025
df_2025_api = get_race_results(2025)

if not df_2025_api.empty:
    df_2025 = df_2025_api
    print(f"Dados 2025 via API: {len(df_2025)} registros")
else:
    print("API sem dados completos de 2025. Usando base manual validada.")
    print("Confirme pontos em: https://www.formula1.com/en/results/2025")

    df_2025 = pd.DataFrame({
        'year':                    [2025] * 20,
        'round':                   [1] * 20,
        'race_name':               ['Australian Grand Prix'] * 20,
        'circuit_name':            ['Albert Park Circuit'] * 20,
        'country':                 ['Australia'] * 20,
        'date':                    ['2025-03-16'] * 20,
        'driver_name': [
            'Lando Norris','Max Verstappen','Oscar Piastri','Charles Leclerc',
            'Lewis Hamilton','George Russell','Carlos Sainz','Fernando Alonso',
            'Yuki Tsunoda','Lance Stroll','Alexander Albon','Nico Hulkenberg',
            'Pierre Gasly','Esteban Ocon','Kevin Magnussen','Valtteri Bottas',
            'Zhou Guanyu','Logan Sargeant','Oliver Bearman','Gabriel Bortoleto'
        ],
        'constructor_name': [
            'McLaren','Red Bull','McLaren','Ferrari','Ferrari','Mercedes',
            'Williams','Aston Martin','RB','Aston Martin','Williams','Haas',
            'Alpine','Alpine','Haas','Sauber','Sauber','Williams','Haas','Sauber'
        ],
        'driver_id': [
            'norris','max_verstappen','piastri','leclerc','hamilton','russell',
            'sainz','alonso','tsunoda','stroll','albon','hulkenberg',
            'gasly','ocon','kevin_magnussen','bottas','zhou','sargeant',
            'bearman','bortoleto'
        ],
        'constructor_id': [
            'mclaren','red_bull','mclaren','ferrari','ferrari','mercedes',
            'williams','aston_martin','rb','aston_martin','williams','haas',
            'alpine','alpine','haas','sauber','sauber','williams','haas','sauber'
        ],
        'driver_nationality': [
            'British','Dutch','Australian','Monegasque','British','British',
            'Spanish','Spanish','Japanese','Canadian','Thai','German',
            'French','French','Danish','Finnish','Chinese','American',
            'British','Brazilian'
        ],
        # corrigido: nacionalidades reais por equipe (era 'British' * 20)
        'constructor_nationality': [
            'British','Austrian','British','Italian','Italian','British',
            'British','British','Austrian','British','British','American',
            'French','French','American','Finnish','Swiss','American',
            'American','Swiss'
        ],
        'grid':             list(range(1, 21)),
        'position':         list(range(1, 21)),
        'position_text':    [str(i) for i in range(1, 21)],
        'points':           [25,18,15,12,10,8,6,4,2,1] + [0]*10,
        'laps':             [58] * 20,
        'status':           ['Finished']*15 + ['Lapped']*3 + ['Retired']*2,
        'fastest_lap_time': [None] * 20,
        'fastest_lap_rank': [None] * 20
    })

# Treinos 2026
print("\nCarregando treinos 2026...")

bahrain_test2 = pd.DataFrame({
    'year': [2026]*22, 'round': [0]*22,
    'race_name':    ['Pre-Season Testing Bahrain']*22,
    'circuit_name': ['Bahrain International Circuit']*22,
    'country':      ['Bahrain']*22,
    'session':      ['Bahrain Test 2']*22,
    'driver_name': [
        'Charles Leclerc','Kimi Antonelli','Oscar Piastri','Lando Norris',
        'Max Verstappen','George Russell','Lewis Hamilton','Pierre Gasly',
        'Oliver Bearman','Gabriel Bortoleto','Franco Colapinto','Nico Hulkenberg',
        'Arvid Lindblad','Esteban Ocon','Isack Hadjar','Carlos Sainz',
        'Liam Lawson','Alexander Albon','Valtteri Bottas','Sergio Perez',
        'Lance Stroll','Fernando Alonso'
    ],
    'constructor_name': [
        'Ferrari','Mercedes','McLaren','McLaren','Red Bull','Mercedes',
        'Ferrari','Alpine','Haas','Audi','Alpine','Audi',
        'Racing Bulls','Haas','Red Bull','Williams',
        'Racing Bulls','Williams','Cadillac','Cadillac',
        'Aston Martin','Aston Martin'
    ],
    'position':     list(range(1,23)),
    'gap_to_leader':[
        0.000, 0.811, 0.869, 0.879,
        1.117, 1.205, 1.416, 1.429,
        1.495, 1.763, 1.826, 1.995,
        2.157, 2.209, 2.268, 2.350,
        2.540, 2.563, 3.298, 3.377,
        3.982, 4.544
    ],
    'total_laps':   [202,197,222,173,204,235,122,179,199,171,180,186,
                     240,205,125,196,167,172,131,135,32,96],
    'points':       [0]*22,
    'status':       ['Testing']*22
})

fp1_australia = pd.DataFrame({
    'year': [2026]*22, 'round': [1]*22,
    'race_name':    ['Australian Grand Prix']*22,
    'circuit_name': ['Albert Park Circuit']*22,
    'country':      ['Australia']*22,
    'session':      ['FP1']*22,
    'driver_name': [
        'Charles Leclerc','Lewis Hamilton','Max Verstappen','Isack Hadjar',
        'Arvid Lindblad','Oscar Piastri','George Russell','Kimi Antonelli',
        'Gabriel Bortoleto','Nico Hulkenberg','Esteban Ocon','Carlos Sainz',
        'Liam Lawson','Oliver Bearman','Alexander Albon','Franco Colapinto',
        'Valtteri Bottas','Pierre Gasly','Lando Norris','Sergio Perez',
        'Lance Stroll','Fernando Alonso'
    ],
    'constructor_name': [
        'Ferrari','Ferrari','Red Bull','Red Bull','Racing Bulls','McLaren',
        'Mercedes','Mercedes','Audi','Audi','Haas','Williams',
        'Racing Bulls','Haas','Williams','Alpine',
        'Cadillac','Alpine','McLaren','Cadillac',
        'Aston Martin','Aston Martin'
    ],
    'position':     list(range(1,23)),
    'gap_to_leader':[
        0.000, 0.469, 0.522, 0.820,
        1.046, 1.075, 1.104, 1.109,
        1.429, 1.702, 1.894, 2.056,
        2.346, 2.415, 2.863, 3.058,
        3.755, 3.768, 4.124, 4.353,
        30.067, None
    ],
    'total_laps': [45]*22,
    'points':     [0]*22,
    'status':     ['Practice']*22
})

fp2_australia = pd.DataFrame({
    'year': [2026]*22, 'round': [1]*22,
    'race_name':    ['Australian Grand Prix']*22,
    'circuit_name': ['Albert Park Circuit']*22,
    'country':      ['Australia']*22,
    'session':      ['FP2']*22,
    'driver_name': [
        'Oscar Piastri','Kimi Antonelli','George Russell','Lewis Hamilton',
        'Charles Leclerc','Max Verstappen','Lando Norris','Arvid Lindblad',
        'Isack Hadjar','Esteban Ocon','Oliver Bearman','Nico Hulkenberg',
        'Liam Lawson','Gabriel Bortoleto','Alexander Albon','Pierre Gasly',
        'Carlos Sainz','Franco Colapinto','Valtteri Bottas','Fernando Alonso',
        'Lance Stroll','Sergio Perez'
    ],
    'constructor_name': [
        'McLaren','Mercedes','Mercedes','Ferrari','Ferrari','Red Bull',
        'McLaren','Racing Bulls','Red Bull','Haas','Haas','Audi',
        'Racing Bulls','Audi','Williams','Alpine',
        'Williams','Alpine','Cadillac','Aston Martin',
        'Aston Martin','Cadillac'
    ],
    'position':     list(range(1,23)),
    'gap_to_leader':[
        0.000, 0.214, 0.320, 0.321,
        0.562, 0.637, 1.065, 1.193,
        1.212, 1.450, 1.597, 1.622,
        1.629, 1.939, 2.118, 2.438,
        2.524, 2.890, 3.931, 4.933,
        6.087, None
    ],
    'total_laps': [55]*22,
    'points':     [0]*22,
    'status':     ['Practice']*22
})

df_2026 = pd.concat([bahrain_test2, fp1_australia, fp2_australia], ignore_index=True)
print(f"Treinos 2026: {len(df_2026)} registros ({df_2026['session'].nunique()} sessoes)")

# Consolidar historico + 2025
cols_comuns = ['year','round','race_name','circuit_name','country','date',
               'driver_id','driver_name','driver_nationality',
               'constructor_id','constructor_name','constructor_nationality',
               'grid','position','position_text','points','laps','status',
               'fastest_lap_time','fastest_lap_rank']

cols_ok = [c for c in cols_comuns
           if c in df_historical.columns and c in df_2025.columns]

df_full = pd.concat([
    df_historical[cols_ok],
    df_2025[cols_ok]
], ignore_index=True)

df_full['position'] = pd.to_numeric(df_full['position'], errors='coerce')
df_full['points']   = pd.to_numeric(df_full['points'],   errors='coerce')

winners = df_full[df_full['position'] == 1].copy()
podiums = df_full[df_full['position'].isin([1,2,3])].copy()

print(f"\nDataset completo (2014-2025):")
print(f"   Registros: {len(df_full):,}")
print(f"   Vitorias:  {len(winners)}")
print(f"   Podios:    {len(podiums)}")

# -----------------------------------------------------------------------------
# BLOCO 3: GABARITO OFICIAL DOS CAMPEOES
# -----------------------------------------------------------------------------

print("\n" + "="*80)
print("BLOCO 3: GABARITO OFICIAL DOS CAMPEOES")
print("="*80)

# Dados confirmados via formula1.com/en/results
CAMPEOES_REAIS = {
    2014: ('Lewis Hamilton',  'Mercedes',  384),
    2015: ('Lewis Hamilton',  'Mercedes',  381),
    2016: ('Nico Rosberg',    'Mercedes',  385),
    2017: ('Lewis Hamilton',  'Mercedes',  363),
    2018: ('Lewis Hamilton',  'Mercedes',  408),
    2019: ('Lewis Hamilton',  'Mercedes',  413),
    2020: ('Lewis Hamilton',  'Mercedes',  347),
    2021: ('Max Verstappen',  'Red Bull',  395),
    2022: ('Max Verstappen',  'Red Bull',  454),
    2023: ('Max Verstappen',  'Red Bull',  575),
    2024: ('Max Verstappen',  'Red Bull',  437),
    2025: ('Lando Norris',    'McLaren',   423),
}

CAMPEOES_CONSTRUTORES = {
    2014: ('Mercedes',  701),
    2015: ('Mercedes',  703),
    2016: ('Mercedes',  765),
    2017: ('Mercedes',  668),
    2018: ('Mercedes',  655),
    2019: ('Mercedes',  739),
    2020: ('Mercedes',  573),
    2021: ('Mercedes',  613),
    2022: ('Red Bull',  759),
    2023: ('Red Bull',  860),
    2024: ('McLaren',   666),
    2025: ('McLaren',   833),
}

print(f"\n{'Ano':<6} {'Piloto Campeao':<25} {'Equipe':<15} {'Pts':<8} {'Construtor':<12} {'Pts'}")
print("-"*75)
for year in sorted(CAMPEOES_REAIS.keys()):
    d, td, pd_ = CAMPEOES_REAIS[year]
    tc, pc     = CAMPEOES_CONSTRUTORES[year]
    pd_s = str(pd_) if pd_ else "N/A"
    pc_s = str(pc)  if pc  else "N/A"
    print(f"{year:<6} {d:<25} {td:<15} {pd_s:<8} {tc:<12} {pc_s}")

# -----------------------------------------------------------------------------
# BLOCO 4: ANALISES HISTORICAS (2014-2025)
# -----------------------------------------------------------------------------

print("\n" + "="*80)
print("BLOCO 4: ANALISES HISTORICAS 2014-2025")
print("="*80)

# 4.1 Heatmap vitorias por equipe/ano
print("\n4.1 HEATMAP — VITORIAS POR EQUIPE E ANO")
wins_hm = winners.groupby(['constructor_name','year']).size().reset_index(name='wins')
hm_data = wins_hm.pivot(index='constructor_name', columns='year', values='wins').fillna(0)
hm_data = hm_data[hm_data.sum(axis=1) > 0]

plt.figure(figsize=(16,6))
sns.heatmap(hm_data, annot=True, cmap='YlOrRd', fmt='g',
            cbar_kws={'label': 'Vitorias'})
plt.title('Vitorias por Equipe e Ano (2014-2025)', fontsize=16, pad=20)
plt.tight_layout()
plt.show()

# 4.2 Evolucao de pontos top 8 equipes
print("\n4.2 EVOLUCAO DE PONTOS — TOP 8 EQUIPES")
team_pts_year = df_full.groupby(['year','constructor_name'])['points'].sum().reset_index()
top8 = df_full.groupby('constructor_name')['points'].sum().sort_values(ascending=False).head(8).index
team_evol = team_pts_year[team_pts_year['constructor_name'].isin(top8)]

fig = px.line(team_evol, x='year', y='points', color='constructor_name',
              title='Evolucao de Pontos por Equipe (2014-2025)',
              markers=True, line_shape='spline')
fig.update_layout(width=1200, height=600)
fig.update_traces(line=dict(width=3), marker=dict(size=8))
fig.show()

# 4.3 Dominancia por ano
print("\n4.3 DOMINANCIA — % DE VITORIAS POR ANO")
dominance_rows = []
for year in sorted(df_full['year'].unique()):
    y_wins  = winners[winners['year'] == year]
    y_races = df_full[df_full['year'] == year]['round'].nunique()
    if y_races > 0 and len(y_wins) > 0:
        team_wins    = y_wins['constructor_name'].value_counts()
        driver_wins  = y_wins['driver_name'].value_counts()
        dominance_rows.append({
            'year':              year,
            'total_races':       y_races,
            'team_dominance':    round(team_wins.iloc[0]   / y_races * 100, 1),
            'driver_dominance':  round(driver_wins.iloc[0] / y_races * 100, 1),
            'unique_winners':    y_wins['driver_name'].nunique(),
            'leading_team':      team_wins.index[0],
            'leading_driver':    driver_wins.index[0]
        })

dominance_df = pd.DataFrame(dominance_rows)

fig = make_subplots(rows=2, cols=1,
                    subplot_titles=['Dominancia da Equipe Lider (%)',
                                    'Vencedores Unicos por Ano'])
fig.add_trace(go.Bar(x=dominance_df['year'], y=dominance_df['team_dominance'],
                     name='Dominancia', marker_color='crimson',
                     text=[f"{d}%<br>{t}" for d,t in zip(
                         dominance_df['team_dominance'], dominance_df['leading_team'])],
                     textposition='outside'), row=1, col=1)
fig.add_trace(go.Bar(x=dominance_df['year'], y=dominance_df['unique_winners'],
                     name='Vencedores Unicos', marker_color='steelblue',
                     text=dominance_df['unique_winners'],
                     textposition='outside'), row=2, col=1)
fig.update_layout(height=700, title_text='Analise de Dominancia por Ano (2014-2025)',
                  title_font_size=18, showlegend=False)
fig.show()

print(f"\n{'Ano':<6} {'Equipe Lider':<16} {'Dom.%':<8} {'Piloto Lider':<22} {'Dom.%':<8} {'Unicos'}")
print("-"*70)
for _, r in dominance_df.iterrows():
    print(f"{int(r['year']):<6} {r['leading_team']:<16} {r['team_dominance']:<8} "
          f"{r['leading_driver']:<22} {r['driver_dominance']:<8} {r['unique_winners']}")

# 4.4 Performance detalhada dos pilotos
print("\n4.4 PERFORMANCE DETALHADA DOS PILOTOS")
driver_stats = df_full.groupby('driver_name').agg(
    total_points  = ('points',   'sum'),
    avg_points    = ('points',   'mean'),
    points_std    = ('points',   'std'),
    avg_position  = ('position', 'mean'),
    pos_std       = ('position', 'std'),
    best_position = ('position', 'min'),
    races_count   = ('round',    'count')
).round(2).reset_index()

driver_stats = driver_stats[driver_stats['races_count'] >= 30]
driver_stats['consistency'] = (1 / (driver_stats['pos_std'] + 0.1)).round(2)

fig = px.scatter(driver_stats,
                 x='avg_position', y='total_points',
                 size='races_count', color='consistency',
                 hover_name='driver_name',
                 title='Pilotos: Posicao Media vs Pontos Totais vs Consistencia (2014-2025)',
                 color_continuous_scale='viridis')
fig.update_xaxes(autorange='reversed')
fig.update_layout(width=1200, height=700)
fig.show()

# corrigido: uso de enumerate para ranking correto (era get_loc, que podia errar)
print(f"\nTOP 10 PILOTOS:")
print(f"{'#':<4}{'Piloto':<22}{'Pts Totais':<12}{'Pos. Media':<12}{'Melhor':<9}{'Corridas':<11}{'Consist.'}")
print("-"*80)
for rank, (_, row) in enumerate(
    driver_stats.sort_values('total_points', ascending=False).head(10).iterrows(), 1
):
    print(f"{rank:<4}{row['driver_name']:<22}{row['total_points']:<12.0f}"
          f"{row['avg_position']:<12.1f}{row['best_position']:<9.0f}"
          f"{row['races_count']:<11.0f}{row['consistency']:<.2f}")

# 4.5 Analise de podios
print("\n4.5 ANALISE DE PODIOS")
podium_bkd = podiums.groupby(['driver_name','position']).size().unstack(fill_value=0)
for col in [1, 2, 3]:
    if col not in podium_bkd.columns:
        podium_bkd[col] = 0
podium_bkd.columns = ['Vitorias','2o Lugar','3o Lugar']
podium_bkd['Total'] = podium_bkd.sum(axis=1)
podium_bkd = podium_bkd[podium_bkd['Total'] >= 5].sort_values('Total', ascending=False)

fig = go.Figure()
fig.add_trace(go.Bar(name='1o - Vitorias', x=podium_bkd.index,
                     y=podium_bkd['Vitorias'], marker_color='gold',
                     text=podium_bkd['Vitorias'], textposition='inside'))
fig.add_trace(go.Bar(name='2o Lugar', x=podium_bkd.index,
                     y=podium_bkd['2o Lugar'], marker_color='silver',
                     text=podium_bkd['2o Lugar'], textposition='inside'))
fig.add_trace(go.Bar(name='3o Lugar', x=podium_bkd.index,
                     y=podium_bkd['3o Lugar'], marker_color='#CD7F32',
                     text=podium_bkd['3o Lugar'], textposition='inside'))
fig.update_layout(title='Distribuicao de Podios por Piloto (2014-2025)',
                  barmode='stack', width=1400, height=600, title_font_size=18)
fig.update_xaxes(tickangle=45)
fig.show()

# 4.6 Pontuacao dos campeoes
print("\n4.6 PONTUACAO DOS CAMPEOES POR ANO")
# Pontos 2025 confirmados via formula1.com: Norris 423 | Verstappen 421 | Piastri 410
champ_data = [{'year': y, 'driver': d, 'team': t, 'points': p}
              for y,(d,t,p) in CAMPEOES_REAIS.items()
              if p is not None and y >= 2014]
champ_df = pd.DataFrame(champ_data)

era_colors = {
    'Lewis Hamilton': '#00D2BE',
    'Nico Rosberg':   '#00D2BE',
    'Max Verstappen': '#1E3A5F',
    'Lando Norris':   '#FF8000'
}

fig = go.Figure()
for driver in champ_df['driver'].unique():
    d = champ_df[champ_df['driver'] == driver]
    fig.add_trace(go.Bar(
        x=d['year'], y=d['points'], name=driver,
        marker_color=era_colors.get(driver, 'gray'),
        text=[f"{driver}<br>{p}pts" for p in d['points']],
        textposition='inside'
    ))
fig.update_layout(title='Pontuacao dos Campeoes da F1 (2014-2025)',
                  xaxis_title='Ano', yaxis_title='Pontos',
                  barmode='stack', width=1200, height=600)
fig.show()

# -----------------------------------------------------------------------------
# BLOCO 5: ANALISE ESPECIAL — GABRIEL BORTOLETO + AUDI 2026
# -----------------------------------------------------------------------------

print("\n" + "="*80)
print("BLOCO 5: ANALISE ESPECIAL — GABRIEL BORTOLETO + AUDI 2026")
print("="*80)

bortoleto_2026 = df_2026[df_2026['driver_name'] == 'Gabriel Bortoleto'].copy()
audi_2026      = df_2026[df_2026['constructor_name'] == 'Audi'].copy()
sessions_order = ['Bahrain Test 2','FP1','FP2']

# Evolucao sessao a sessao
print(f"\nBORTOLETO — EVOLUCAO NOS TREINOS 2026:")
print(f"{'Sessao':<22} {'Posicao':<10} {'Gap':<12} {'Voltas'}")
print("-"*55)
for s in sessions_order:
    sd = bortoleto_2026[bortoleto_2026['session'] == s]
    if len(sd) > 0:
        pos  = int(sd['position'].iloc[0])
        gap  = sd['gap_to_leader'].iloc[0]
        laps = sd.get('total_laps', pd.Series([None])).iloc[0]
        gap_s = f"+{gap:.3f}s" if pd.notna(gap) else "N/A"
        print(f"{s:<22} P{pos:<9} {gap_s:<12} {laps}")

# Tendencia
positions_bort = []
for s in sessions_order:
    sd = bortoleto_2026[bortoleto_2026['session'] == s]
    if len(sd) > 0:
        positions_bort.append(int(sd['position'].iloc[0]))

if len(positions_bort) >= 2:
    trend = positions_bort[-1] - positions_bort[0]
    print(f"\nTendencia: {positions_bort[0]}o -> {positions_bort[-1]}o", end=" ")
    if trend < 0:
        print(f"(melhorou {abs(trend)} posicoes)")
    elif trend > 0:
        print(f"(caiu {trend} posicoes)")
    else:
        print("(estavel)")

# Comparativo Bortoleto x Hulkenberg
print(f"\nBORTOLETO vs HULKENBERG (dentro da Audi):")
print(f"{'Sessao':<22} {'Bortoleto':<18} {'Hulkenberg':<18} {'Resultado'}")
print("-"*75)

bort_wins = 0
hulk_wins = 0
for s in sessions_order:
    b = df_2026[(df_2026['driver_name']  == 'Gabriel Bortoleto') & (df_2026['session'] == s)]
    h = df_2026[(df_2026['driver_name']  == 'Nico Hulkenberg')   & (df_2026['session'] == s)]
    if len(b) > 0 and len(h) > 0:
        pb   = int(b['position'].iloc[0])
        ph   = int(h['position'].iloc[0])
        gb   = b['gap_to_leader'].iloc[0]
        gh   = h['gap_to_leader'].iloc[0]
        gb_s = f"P{pb} (+{gb:.3f}s)" if pd.notna(gb) else f"P{pb}"
        gh_s = f"P{ph} (+{gh:.3f}s)" if pd.notna(gh) else f"P{ph}"
        if pb < ph:
            resultado = "Bortoleto"
            bort_wins += 1
        else:
            resultado = "Hulkenberg"
            hulk_wins += 1
        print(f"{s:<22} {gb_s:<18} {gh_s:<18} {resultado}")

print(f"\n   Placar: Bortoleto {bort_wins} x {hulk_wins} Hulkenberg")
if bort_wins > hulk_wins:
    print("   Bortoleto vence o duelo interno.")
elif hulk_wins > bort_wins:
    print("   Hulkenberg vence o duelo interno.")
else:
    print("   Duelo empatado.")

# Grafico evolucao Bortoleto vs Hulkenberg
fig = go.Figure()
for driver, color, symbol in [
    ('Gabriel Bortoleto', '#009C3B', 'circle'),
    ('Nico Hulkenberg',   '#FF6B6B', 'square')
]:
    xs, ys = [], []
    for s in sessions_order:
        d = df_2026[(df_2026['driver_name'] == driver) & (df_2026['session'] == s)]
        if len(d) > 0 and pd.notna(d['gap_to_leader'].iloc[0]):
            xs.append(s)
            ys.append(d['gap_to_leader'].iloc[0])
    fig.add_trace(go.Scatter(
        x=xs, y=ys,
        mode='lines+markers+text',
        name=driver,
        line=dict(color=color, width=3),
        marker=dict(size=12, symbol=symbol),
        text=[f"+{g:.3f}s" for g in ys],
        textposition='top center'
    ))

fig.update_layout(
    title='Bortoleto vs Hulkenberg — Evolucao nos Treinos 2026',
    xaxis_title='Sessao',
    yaxis_title='Gap para o Lider (s) — menor e melhor',
    yaxis=dict(autorange='reversed'),
    width=1000, height=500
)
fig.show()

# Posicionamento Audi vs campo (FP2)
print(f"\nPOSICIONAMENTO AUDI vs CAMPO (FP2 — mais recente):")
fp2 = df_2026[df_2026['session'] == 'FP2'].copy()
audi_team_gaps = fp2.groupby('constructor_name')['gap_to_leader'].min().sort_values()

for i,(team,gap) in enumerate(audi_team_gaps.items(), 1):
    if pd.notna(gap):
        hl  = " <- Audi (BR)" if team == 'Audi' else ""
        bar = "█" * max(1, int((5-gap)*4))
        print(f"{i:2d}. {team:<16}: +{gap:.3f}s  {bar}{hl}")

# -----------------------------------------------------------------------------
# BLOCO 6: PROBABILIDADE E PREVISAO FIM DE TEMPORADA 2026
# -----------------------------------------------------------------------------

print("\n" + "="*80)
print("BLOCO 6: PREVISAO DE CLASSIFICACAO FINAL 2026")
print("="*80)

session_weights = {'FP2': 3, 'FP1': 2, 'Bahrain Test 2': 1}

perf_rows = []
for driver in df_2026['driver_name'].unique():
    dd   = df_2026[df_2026['driver_name'] == driver]
    team = dd['constructor_name'].iloc[0]
    wsum = 0
    wtot = 0
    for _, row in dd.iterrows():
        g = row.get('gap_to_leader')
        if g is not None and pd.notna(g):
            w     = session_weights.get(row['session'], 1)
            wsum += g * w
            wtot += w
    avg_gap = wsum / wtot if wtot > 0 else 9.999
    perf_rows.append({'driver_name': driver,
                      'constructor_name': team,
                      'weighted_gap': round(avg_gap, 3)})

perf_df       = pd.DataFrame(perf_rows)
max_gap_p     = perf_df['weighted_gap'].max()
perf_df['score_2026'] = ((max_gap_p - perf_df['weighted_gap']) / max_gap_p * 100).round(2)

# Score historico recente (2022-2025)
df_recent   = df_full[df_full['year'] >= 2022].copy()
team_hist   = df_recent.groupby('constructor_name')['points'].sum().reset_index()
team_hist.columns = ['constructor_name','hist_points']
max_hist    = team_hist['hist_points'].max()
team_hist['score_historico'] = (team_hist['hist_points'] / max_hist * 100).round(2)

# Merge score 2026 + historico
team_s26    = perf_df.groupby('constructor_name')['score_2026'].max().reset_index()
pred_df     = team_s26.merge(team_hist, on='constructor_name', how='left')
pred_df['score_historico'] = pred_df['score_historico'].fillna(0)

# Score final: 60% treinos 2026 + 40% historico
pred_df['score_final'] = (pred_df['score_2026'] * 0.60 +
                           pred_df['score_historico'] * 0.40).round(2)

total_score  = pred_df['score_final'].sum()
pred_df['probabilidade'] = (pred_df['score_final'] / total_score * 100).round(1)
pred_df = pred_df.sort_values('probabilidade', ascending=False).reset_index(drop=True)
pred_df['ranking'] = range(1, len(pred_df)+1)

# Pontos previstos (base ~24 corridas)
TOTAL_PTS = 4200
pred_df['pontos_previstos'] = (pred_df['probabilidade'] / 100 * TOTAL_PTS).round(0).astype(int)

# Ranking de construtores
print(f"\nRANKING PREVISTO — CONSTRUTORES 2026:")
print(f"{'#':<4}{'Equipe':<18}{'Score 26':<11}{'Score Hist':<13}{'Score Final':<13}{'Prob.':<9}{'Pts Prev.'}")
print("-"*80)
medals = ['1o.','2o.','3o.']
for i, row in pred_df.iterrows():
    m  = medals[i] if i < 3 else f"  {i+1}."
    hl = " (BR)" if row['constructor_name'] == 'Audi' else ""
    print(f"{m:<4}{row['constructor_name']:<18}{row['score_2026']:<11.1f}"
          f"{row['score_historico']:<13.1f}{row['score_final']:<13.1f}"
          f"{row['probabilidade']:>5.1f}%   {row['pontos_previstos']}{hl}")

# Ranking de pilotos
max_gap_d     = perf_df['weighted_gap'].max()
perf_df['score_d'] = ((max_gap_d - perf_df['weighted_gap']) / max_gap_d * 100)
tot_d         = perf_df['score_d'].sum()
perf_df['prob_d']  = (perf_df['score_d'] / tot_d * 100).round(1)
perf_df['pts_d']   = (perf_df['prob_d'] / 100 * TOTAL_PTS / 2).round(0).astype(int)
perf_df = perf_df.sort_values('prob_d', ascending=False).reset_index(drop=True)
perf_df['rank_d'] = range(1, len(perf_df)+1)

print(f"\nRANKING PREVISTO — PILOTOS 2026:")
print(f"{'#':<4}{'Piloto':<24}{'Equipe':<16}{'Prob.':<9}{'Pts Previstos'}")
print("-"*65)
for i, row in perf_df.iterrows():
    m  = medals[i] if i < 3 else f"  {i+1}."
    hl = " <- BR" if row['driver_name'] == 'Gabriel Bortoleto' else ""
    print(f"{m:<4}{row['driver_name']:<24}{row['constructor_name']:<16}"
          f"{row['prob_d']:>5.1f}%   {row['pts_d']}{hl}")

# Grafico final probabilidades
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=['Probabilidade por Equipe (%)',
                                    'Pontos Previstos — Top 10 Pilotos'])

c_teams = ['gold' if i==0 else 'silver' if i==1 else
           '#CD7F32' if i==2 else
           '#009C3B' if t=='Audi' else 'steelblue'
           for i,t in enumerate(pred_df['constructor_name'])]

fig.add_trace(go.Bar(x=pred_df['constructor_name'], y=pred_df['probabilidade'],
                     marker_color=c_teams,
                     text=[f"{p:.1f}%" for p in pred_df['probabilidade']],
                     textposition='outside'), row=1, col=1)

top10_d = perf_df.head(10)
c_drv   = ['#009C3B' if d=='Gabriel Bortoleto' else 'steelblue'
           for d in top10_d['driver_name']]
fig.add_trace(go.Bar(x=top10_d['driver_name'], y=top10_d['pts_d'],
                     marker_color=c_drv,
                     text=top10_d['pts_d'],
                     textposition='outside'), row=1, col=2)

fig.update_layout(title='Previsao Final — F1 2026',
                  height=600, width=1400, showlegend=False)
fig.update_xaxes(tickangle=45)
fig.show()

# -----------------------------------------------------------------------------
# BLOCO 7: INSIGHTS COMPLETOS
# -----------------------------------------------------------------------------

print("\n" + "="*80)
print("BLOCO 7: INSIGHTS COMPLETOS")
print("="*80)

bort_row  = perf_df[perf_df['driver_name'] == 'Gabriel Bortoleto'].iloc[0]
hulk_row  = perf_df[perf_df['driver_name'] == 'Nico Hulkenberg'].iloc[0]
top1_team = pred_df.iloc[0]
top1_drv  = perf_df.iloc[0]

print(f"""
{"-"*76}
INSIGHT 1 — MUDANCA DE ERAS (2014-2025)
{"-"*76}
  - Era Mercedes/Hamilton:   2014-2020 (7 titulos consecutivos)
  - Era Red Bull/Verstappen: 2021-2024 (4 titulos)
  - Era McLaren/Norris:      2025 (quebrou a hegemonia)
  - Regulamentos 2026:       nova era comeca — quem domina?

{"-"*76}
INSIGHT 2 — SURPRESAS 2026 (novos regulamentos)
{"-"*76}
  - Ferrari lidera Bahrain Test 2 e FP1 — candidata ao titulo
  - McLaren reage fortemente no FP2 — defesa do titulo?
  - Mercedes: ritmo de corrida solido, muitas voltas nos testes
  - Red Bull: perdeu lideranca com novos regulamentos
  - Aston Martin: situacao critica — motor Honda com problemas
  - Cadillac e Audi: estreias discretas mas sem catastrofes

{"-"*76}
INSIGHT 3 — GABRIEL BORTOLETO
{"-"*76}
  - Ranking previsto no campeonato de pilotos: {int(bort_row['rank_d'])}o
  - Pontos estimados na temporada: ~{int(bort_row['pts_d'])} pontos
  - Duelo interno vs Hulkenberg: Bortoleto {bort_wins}x{hulk_wins} Hulkenberg
  - Como estreante, desempenho ja supera expectativas
  - Audi tende a melhorar ao longo do ano (motor novo, updates)
  - Potencial de brigar por pontos consistentes a partir do 2o semestre

{"-"*76}
INSIGHT 4 — PREVISAO FINAL 2026
{"-"*76}
  1o. Equipe favorita: {pred_df.iloc[0]['constructor_name']} ({pred_df.iloc[0]['probabilidade']:.1f}%)
  2o. 2a favorita:     {pred_df.iloc[1]['constructor_name']} ({pred_df.iloc[1]['probabilidade']:.1f}%)
  3o. 3a favorita:     {pred_df.iloc[2]['constructor_name']} ({pred_df.iloc[2]['probabilidade']:.1f}%)

  1o. Piloto favorito: {perf_df.iloc[0]['driver_name']} ({perf_df.iloc[0]['prob_d']:.1f}%)
  2o. 2o favorito:     {perf_df.iloc[1]['driver_name']} ({perf_df.iloc[1]['prob_d']:.1f}%)
  3o. 3o favorito:     {perf_df.iloc[2]['driver_name']} ({perf_df.iloc[2]['prob_d']:.1f}%)
""")

print("\n" + "="*80)
print("CODIGO COMPLETO — VALIDACAO CONCLUIDA.")
print("   8 Blocos | 6 Analises historicas | Bortoleto (BR) | Previsao 2026")
print("="*80)
```
